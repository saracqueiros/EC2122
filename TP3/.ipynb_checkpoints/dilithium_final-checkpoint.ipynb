### Esquemas pós-quânticos de assinaturas digitais - Dilithium
Este esquema de assinatura digital baseia-se na dificuldade de encontrar vetores de pequenas dimensões em *lattices* e tem os seguintes principios como critério:

* Implementação segura e fácil;
* Conversadora em relação aos parâmetros utilizados - uma vez que o objetivo é utilizá-los a longo prazo e facilitar a implementação futura de algoritmos quânticos na mesma ordem espacial que os algoritmos atuais.
* Minimizar o tamanho da chave pública + assinatura - devido à geral transmissão de ambas nas aplicações.
* Modularidade - Fácil variação face à segurança implementada.


De forma a implementar este esquema é necessário perceber que existem 3 passos principais:
1. Geração das chaves
2. Assinar a mensagem 
3. Verificar a assinatura 



```python
'''self.n = 256
self.q = 2^23 - 2^13 + 1
self.k = 8
self.l = 7
self.eta = 2
self.tau = 60
self.beta = 120
self.gama1 = 2^19
self.gama2 = (self.q)-1/32

Zx.<x> = ZZ[]
Zq.<z> = PolynomialRing(GF(self.q))
self.Rq = QuotientRing(Zq,z^self.n+1)
self.R = QuotientRing(Zx, x^self.n+1)

self.Mr  = MatrixSpace(self.Rq,self.k,self.l)
  '''

q = 2^23 - 2^13 + 1
tau = 60


def highBits(r, alfa):
        (r1,r0) = decompose(r, alfa)
        return r1

def lowBits( r, alfa):
    (r1,r0) = decompose(r, alfa)
    return r0

def decompose(r, alfa):
    r = mod(r, q)
    r0 = int(mod(r,int(alfa)))
    if (r-r0 == q-1):
        r1 = 0
        r0 = r0-1
    else:
        r1 = (r-r0)/int(alfa)
    return (r1,r0)

def highBitsPoli(poli,alfa):
    k = poli.list()
    for i in range(len(k)):
        h = k[i]
        h = h.list()
        for j in range(len(h)):
            h[j]=highBits(int(h[j]), alfa)
        k[i]=h
    return k

def lbPoli(poli,alfa):
    k = poli.list()
    for i in range(len(k)):
        h = k[i]
        h = h.list()
        for j in range(len(h)):
            h[j] = lowBits(int(h[j]),alfa)
        k[i] = h
    return k

def access_bit(data, num):                              #Passa de Bytes para bits
    base = int(num // 8)
    shift = int(num % 8)
    return (data[base] & (1<<shift)) >> shift

def SampleInBall(r):
    sl = [access_bit(r[:8],i) for i in range(len(r[:8])*8)]
    k = 8 # descartar primeiros 8
    c = [0] * 256 

    for i in range (256-tau, 256):
        while (int(r[k])>i):
            k +=1 

        j = int(r[k])
        k += 1
        s = int(sl[i-196])

        c[i] = c[j]
        c[j] = (-1)^(s)
    return c

def Shake(a,b):
    shake = hashlib.shake_256()
    shake.update(a)
    shake.update(b)
    s = shake.digest(int(256))
    return s

def Hash(a,b):
    r = Shake(a,b)
    c = SampleInBall(r)
    return c

def norma_infinito(pol,n):
    J = pol.list()
    for i in range(len(J)):
        k = J[i]
        K = k.list()
        for j in range(len(K)):
            K[j] = abs(int(K[j]))
        J[i] = K
    L = []
    for i in range(len(J)):
        L.append(max(J[i]))
    return max(L)

def norma_inf_vet(vetor):
    for i in range(vetor.nrows()):
        norm = norma_infinito(vetor[i],self.q)
        vetor[i] = norm
    return max(vetor)


def norma_inf_matriz(matriz):
    L = []
    for i in range(len(matriz)):
        k = matriz[i]
        for j in range(len(k)):
            if k[j] < 0:
                k[j] = abs(k[j])
            L.append(max(k))
    for i in range(len(L)):
        J = []
        J.append(max(L))
    return J[0]
```

```python
import os
import math
import sys
import hashlib
import numpy as np

class DILIT:
    #Geração dos parâmetros para Segurança Nível 5
    def __init__(self):                                           
        self.n = 256
        self.q = 2^23 - 2^13 + 1
        self.k = 8
        self.l = 7
        self.eta = 2
        self.tau = 60
        self.beta = 120
        self.gama1 = 2^19
        self.gama2 = (self.q)-1/32
        
        Zx.<x> = ZZ[]
        Zq.<z> = PolynomialRing(GF(self.q))
        self.Rq = QuotientRing(Zq,z^self.n+1)
        self.R = QuotientRing(Zx, x^self.n+1)
    
        self.Mr  = MatrixSpace(self.Rq,self.k,self.l)
        
        
        
     #Geração a matriz A em Rq
    def genA(self):
        matrix = []
        for i in range(self.k*self.l):
            matrix.append(self.Rq.random_element())
        return self.Mr(matrix)
         
    
     
    #Gerar vetores S em Rq 
    def genS(self, limit, tam):
        vetor = MatrixSpace(self.Rq,tam,1)
        matrix = []
        for i in range(tam):
            polinom = []
            for j in range(self.n):
                polinom.append(randint(1,limit))
            matrix.append(self.Rq(polinom))
        return vetor(matrix)
        
    
    def keygen(self):
        # 1. Criação da matriz A: A←Rk×l
        A = self.genA()
        
        # 2. Criação dos vetores s1 e s2
        #(s1,s2)←Sηl ×Sηk
        s1 = self.genS(self.eta, self.l)
        s2 = self.genS(self.eta, self.k)
        
        #3.Cálculo de t := As1 + s2
        t = A*s1 + s2
        
        #4. return(pk=(A,t),sk=(A,t,s1,s2))
        pubkey = (A,t)
        privkey = (A,t,s1,s2)
        
        return pubkey, privkey
    
    def sign(self, seckey, M): 
        #Extrair os valores da chave privada 
        A, t, s1, s2 = seckey
        vetor = MatrixSpace(self.Rq,self.k,1)
        #1. z := ⊥
        z = 0
        #2. while z = ⊥ do
        while(z==0):
            # 3. Criação do vetor y a partir do genS
            y = self.genS(int(self.gama1-1) , self.l)
            #4. w1 :=HighBits(Ay,2γ)
                # Cálculo Ay
            Ay = A* y
                # Aplicar HighBits
            w1 = str(highBitsPoli(Ay, 2*self.gama2)).encode()
            
            k = M.encode()
            
            # 5. c €Bt:= H(M∥w1)   Hash de M || w1
            c = Hash(k, w1)
            cq = self.Rq(c)
            
            # 6. z := y + cs1 --> Cálculo de z
            z = y + cq*s1
            
            #7. if ∥z∥∞ ≥γ1−βor∥LowBits(Ay−cs2,2γ2)∥∞ ≥γ2−β,thenz:=⊥
            if self.norma_inf_vet(z)[0]>= self.gama1 - self.beta and self.norma_inf_matriz(self.lbPoli(A*y-cq*s2,2*self.gama2))>= self.gama2-self.beta:
                sigma = (z,c)
                return sigma
            else:
                z=0
                
            #return σ=(z,c)
            
            
    # Função para verificar uma assinatura
    def verify(self,pk, M, sigma):
        #Extrair a chave pública 
        A,t = pk
        z,c = sigma
        
        cq = self.Rq(c)
        
        #1. w1′ := HighBits(A*z − c*t, 2*γ2)
        w1 = self.highBitsPoli(A*z - cq*t, 2*self.gama2)
       
        w1_ = str(w1).encode()
        k = M.encode()
        c_ = self.Hash(k,w1_)
        
        #if return [∥z∥∞ <γ1 −β and c=H(M ∥w1′)]
  
        if c_ == c:
            print("Assinatura Válida!")
        else:
            print("Assinatura Inválida!")
        
```
